```python
import math

# sigmoide:
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

```python
# relu:
import torch

class ReLU(torch.nn.Module): # importa as funcionlidades de uma rede neural básica
    def __init__(self):
        super().__init__()
    
    def forward(self, x):
        return torch.max(torch.zeros_like(x), x) # escolhe o valor maior, entre 0 e x[i], para todo x[i]
```

```python
# mean squared error
import numpy as np

def mse_loss(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)
```

```python
# binary cross entropy

import numpy as np

def cross_entropy_loss(y_true, y_pred):
    epsilon = 1e-15
    y_pred = np.clip(y_pred, epsilon, 1. - epsilon) # para evitar divisões por 0
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
```
