```python
import cv2 as cv # importa a open cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread('/kaggle/working/ab6761610000e5eb31f6ab67e6025de876475814')
```

```python
plt.imshow(img)
```

```python
# convertendo para rgb
rgb_img = cv.cvtColor(img, cv.COLOR_BGR2RGB) # converte a imagem de BGR para RGB
plt.imshow(rgb_img)
```

```python
# convertendo para grayscale
gray_img = cv.cvtColor(rgb_img, cv.COLOR_RGB2GRAY)
plt.imshow(gray_img, cmap='gray')
```

```python
# analisando os shapes
rgb_img.shape, gray_img.shape
```

Observa-se que a imagem em escala de cinza não tem a dimensão de canais de cor.

```python
# mudando o tamanho da imagem
gray_img_smll = cv.resize(gray_img, (100, 100))
plt.imshow(gray_img_smll, cmap='gray')
```

```python
# cortando a imagem
# é o mesmo que fazer um slice num array
plt.imshow(gray_img[:300, :], cmap='gray')
```

```python
plt.imshow(gray_img[:, :300], cmap='gray')
```
